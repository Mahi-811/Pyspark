```python
my_ddl_schema = """
Item_identifier STRING,
Item_Weight STRING,
Item_Fat_Content STRING,
Item_Visibility DOUBLE,
Item_Type STRING,
Item_MRP DOUBLE,
Outlet_Identifier STRING,
Outlet_Establishment_Year INTEGER,
Outlet_Size STRING,
Outlet_Location_Type STRING,
Outlet_Type STRING,
Item_Outlet_Sales DOUBLE




"""
```

```python
df = spark.read.format('csv')\
    .schema(my_ddl_schema)\
      .option('header',True)\
        .load('/Volumes/workspace/default/tutorial/BigMart Sales.csv/')
```

```python
df.display()
```

```python
df.printSchema()
```
